```python
Stateagg = df.groupby(['State'])['Price'].agg(['mean', 'median', 'sum'])
```

```python
Stateagg.columns = Stateagg.columns.map(lambda x: 'state_' + str(x) )
```

```python
Stateagg
```

```python
Cityagg = df.groupby(['City'])['Price'].agg(['mean', 'median', 'sum'])
Cityagg.columns = Cityagg.columns.map(lambda x: 'city_' + str(x) )
```

```python
Cityagg
```

```python
Localityagg = df.groupby(['Locality'])['Price'].agg(['mean', 'median', 'sum'])
Localityagg.columns = Localityagg.columns.map(lambda x: 'locality_' + str(x) )
```

```python
Localityagg.head()
```

```python
sf =df.copy()
```

```python
sf =sf.join(Localityagg, on = ['Locality'], how ="inner")
```

```python
sf =sf.join(Cityagg, on = ['City'], how ="inner")
```

```python
sf =sf.join(Stateagg, on = ['State'], how ="inner")
```

```python
sf
```

```python
brandagg = df.groupby('Brand')['Price'].agg(['mean','median','sum'])
brandagg.columns = brandagg.columns.map(lambda x: 'brand_' + str(x) )
```

```python
sf =sf.join(brandagg, on = ['Brand'], how ="inner")
```

```python
target = df.groupby(['Brand','State']).agg(['median','mean','sum']).reset_index()
target
```

```python
columns = ['Brand','State']

# Iterate through the variables names
for var in target.columns.levels[0]:
    # Skip the id name
    if var != 'Brand' and var != 'State':
        
        # Iterate through the stat names
        for stat in target.columns.levels[1][:-1]:
            # Make a new column name for the variable and stat
            columns.append('state%s_%s' % (var, stat))
```

```python
columns
```

```python
target.columns = columns
target.head()
```

```python
ntarget = target[['Brand','State','statePrice_median','statePrice_mean','statePrice_sum']]
```

```python
nsf = sf.merge(ntarget, on = ['Brand','State'] ,how ='inner')
```
